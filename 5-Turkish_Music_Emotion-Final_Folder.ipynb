### Categorization of Music Emotion

## Inputs (Features) for Model (X, Y)

```python
from sklearn.model_selection import train_test_split, cross_val_score

x = main_folder.drop('Class', axis=1)
y = main_folder['Class']
x_train, x_test, y_train, y_test= train_test_split(x, y, test_size=0.2, random_state=5, stratify =y)
```

## Standardization of Inputs (Features)

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(x_train)
x_train_std = scaler.transform(x_train)
x_test_std = scaler.transform(x_test)
```

# MODELLING

## XGBoostingClassifier Modelling : 
### Model Training and Prediction

```python
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, roc_auc_score, precision_score
boost = XGBClassifier(eval_metric='mlogloss', random_state=42)

boost.fit(x_train_std, y_train)
y_pred_boost=boost.predict(x_test_std)

accuracy_boost= accuracy_score(y_test, y_pred_boost)
print ('Boosting Accuracy:', accuracy_boost)
```

## Logistic Regression Modelling :
### Determininig Best Parameters with Grid Search, Model Training, Prediction and Scoring

```python
# Finding Best Parameters with GridSearch
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV

log_model=LogisticRegression(max_iter = 10000, solver='liblinear')

C_log_range = [0.2, 0.3, 0.4, 0.5, 0.8, 1, 1.4, 2, 2.5, 3, 3.5, 4]
penalties=['l1', 'l2']

param_log = {'C': C_log_range, 'penalty': penalties}
grid_search_log = GridSearchCV (estimator=log_model, param_grid=param_log, cv=5, verbose=2, n_jobs=-1)

grid_search_log.fit(x_train_std, y_train)

cv_results_log = pd.DataFrame(grid_search_log.cv_results_)

print("\nBest Accuracy Average Score:", grid_search_log.best_score_)
print("\nBest Parameters:", grid_search_log.best_params_, "\n")
```

### Logistic Regression Model Training with Best Parameters and Prediction 

```python
log_reg_best=LogisticRegression(max_iter = 10000, solver='liblinear', C=0.8, penalty='l1')
log_reg_best.fit(x_train_std, y_train)
y_pred_log=log_reg_best.predict(x_test_std)

accuracy_log= accuracy_score(y_test, y_pred_log)
print ('Logistic Reggresion Accuracy:', accuracy_log)
```

## Support Vector Machine (SVC) Classifier Modelling :
### Model Training, Prediction and Scoring

```python
from sklearn.svm import SVC
svm = SVC(probability=True)

kernel= ['rbf', 'poly', 'sigmoid']
C_svm = [0.01, 0.1, 1, 10, 100]
gamma = [1, 0.1, 0.01, 0.001]
param_svm = {'C': C_svm, 'kernel': kernel, 'gamma': gamma }

grid_search_svm = GridSearchCV(estimator=svm, param_grid=param_svm, scoring='accuracy', cv=5, verbose=2, n_jobs=-1)

grid_search_svm.fit(x_train_std, y_train)

cv_results_svm = pd.DataFrame(grid_search_svm.cv_results_)

print("\nCross-validation Best Accuracy Average Score:", "\n")
print("\nBest Parameters:", grid_search_svm.best_params_, "\n")
```

### Support Vector Machine Model Training with Best Parameters and Prediction

```python
svm_best = SVC(C=100, gamma=0.001, kernel='sigmoid', probability=True)
svm_best.fit(x_train_std, y_train)
y_pred_svm=svm_best.predict(x_test_std)
accuracy_svm= accuracy_score(y_test, y_pred_svm)
print ('SVM (Support Vector Machine) Accuracy:', accuracy_svm)
```
